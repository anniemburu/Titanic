```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from imblearn.over_sampling import RandomOverSampler
```

```python
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Machine Learning Projects/Titanic/train_data.csv")
data.head()
```

```python
# iondependent and dependent variables
X = data.drop(['Survived'], axis = 1)
Y = data['Survived']
```

```python
oversampler = RandomOverSampler(random_state=42)

X_resampled, y_resampled = oversampler.fit_resample(X, Y)
```

```python
y_resampled.value_counts()
```

```python
import lazypredict
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from lazypredict.Supervised import LazyClassifier, LazyRegressor

import warnings
import random
warnings.filterwarnings("ignore")
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_resampled, y_resampled, test_size=0.10, random_state=42) #split dataset
```

## Train Model with Wheels

```python
clf = LazyClassifier(predictions=True)
#clf = LazyClassifier(verbose = 0, )
models, prediction = clf.fit(X_train, X_val, y_train, y_val)
```

```python
models
```

## Modelling


```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.ensemble import GradientBoostingClassifier
import xgboost as xgb
from sklearn.metrics import accuracy_score, roc_auc_score,f1_score
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.linear_model import RidgeClassifierCV
from sklearn.svm import SVC
```

```python
params = {
    "booster" : ("gbtree", 'gblinear',"dart"),
    "learning_rate" : [0.1,0.01,0.001],
    "tree_method" : ('auto', 'exact', 'approx', 'hist', 'gpu_hist'),
    'process_type' : ('default', 'update')
}
```

```python
scaler = StandardScaler()

scaler.fit(X_train)

X_train , X_val = scaler.transform(X_train), scaler.transform(X_val)
```

```python
#Vanilla xgb
svc = xgb.XGBClassifier()
svc.fit(X_train, y_train)

train_pred = svc.predict(X_train)
val_pred = svc.predict(X_val)

print("Train")
print("Accuracy : {} \n".format(accuracy_score(y_train, train_pred)))
print("ROC AUC : {} \n".format(roc_auc_score(y_train, train_pred)))
print("F1 Score : {} \n".format(f1_score(y_train, train_pred)))

print("*****************************************************")

print("Validation")
print("Accuracy : {} \n".format(accuracy_score(y_val, val_pred)))
print("ROC AUC : {} \n".format(roc_auc_score(y_val, val_pred)))
print("F1 Score : {} \n".format(f1_score(y_val, val_pred)))
```

```python
#Vanilla SVM
svc = SVC()
svc.fit(X_train, y_train)

train_pred = svc.predict(X_train)
val_pred = svc.predict(X_val)

print("Train")
print("Accuracy : {} \n".format(accuracy_score(y_train, train_pred)))
print("ROC AUC : {} \n".format(roc_auc_score(y_train, train_pred)))
print("F1 Score : {} \n".format(f1_score(y_train, train_pred)))

print("*****************************************************")

print("Validation")
print("Accuracy : {} \n".format(accuracy_score(y_val, val_pred)))
print("ROC AUC : {} \n".format(roc_auc_score(y_val, val_pred)))
print("F1 Score : {} \n".format(f1_score(y_val, val_pred)))
```

```python
#Vanilla LDA
svc = LDA()
svc.fit(X_train, y_train)

train_pred = svc.predict(X_train)
val_pred = svc.predict(X_val)

print("Train")
print("Accuracy : {} \n".format(accuracy_score(y_train, train_pred)))
print("ROC AUC : {} \n".format(roc_auc_score(y_train, train_pred)))
print("F1 Score : {} \n".format(f1_score(y_train, train_pred)))

print("*****************************************************")

print("Validation")
print("Accuracy : {} \n".format(accuracy_score(y_val, val_pred)))
print("ROC AUC : {} \n".format(roc_auc_score(y_val, val_pred)))
print("F1 Score : {} \n".format(f1_score(y_val, val_pred)))
```

```python
#Vanilla LDA
svc = RidgeClassifierCV()
svc.fit(X_train, y_train)

train_pred = svc.predict(X_train)
val_pred = svc.predict(X_val)

print("Train")
print("Accuracy : {} \n".format(accuracy_score(y_train, train_pred)))
print("ROC AUC : {} \n".format(roc_auc_score(y_train, train_pred)))
print("F1 Score : {} \n".format(f1_score(y_train, train_pred)))

print("*****************************************************")

print("Validation")
print("Accuracy : {} \n".format(accuracy_score(y_val, val_pred)))
print("ROC AUC : {} \n".format(roc_auc_score(y_val, val_pred)))
print("F1 Score : {} \n".format(f1_score(y_val, val_pred)))
```

```python
## Hypermaram tuning
params = {
    "kernel" : ('linear', 'poly', 'rbf', 'sigmoid'),
    'C': [1,2,3,4,5],
    'gamma' : ('scale', 'auto', 'float'),
    'class_weight': (None, 'balanced'),
    'decision_function_shape': ('ovo', 'ovr')
}

svc = SVC()
clf = GridSearchCV(svc, params)
clf.fit(X_train, y_train)

print(clf.best_params_)

train_pred = clf.predict(X_train)
val_pred = clf.predict(X_val)

print("Train")
print("Accuracy : {} \n".format(accuracy_score(y_train, train_pred)))
print("ROC AUC : {} \n".format(roc_auc_score(y_train, train_pred)))
print("F1 Score : {} \n".format(f1_score(y_train, train_pred)))

print("*****************************************************")

print("Validation")
print("Accuracy : {} \n".format(accuracy_score(y_val, val_pred)))
print("ROC AUC : {} \n".format(roc_auc_score(y_val, val_pred)))
print("F1 Score : {} \n".format(f1_score(y_val, val_pred)))
```
