## Cleaning

```python
# deal with null ages
train_mean_age = train.groupby(['Sex'])['Age'].mean()   #mean for each sex train
test_mean_age = test.groupby(['Sex'])['Age'].mean()   #mean for each sex train

#train
train.loc[pd.isnull(train['Age']) & (train['Sex'] == "female"), "Age"] = round(train_mean_age.loc['female'], 2)
train.loc[pd.isnull(train['Age']) & (train['Sex'] == "male"), "Age"] = round(train_mean_age.loc['male'], 2)

#test
test.loc[pd.isnull(test['Age']) & (test['Sex'] == "female"), "Age"] = round(test_mean_age.loc['female'], 2)
test.loc[pd.isnull(test['Age']) & (test['Sex'] == "male"), "Age"] = round(test_mean_age.loc['male'], 2)
```

```python
#change data types
columns = ['Survived', 'Sex', 'Embarked',"Pclass"]

for col in columns:
  train[col] = train[col].astype('string')
  train[col] = train[col].astype('category')

  try:
    test[col] = test[col].astype('string')
    test[col] = test[col].astype('category')
  except:
    pass

train.info()
```

```python
# EDA Ready
eda_train = train.drop(['Cabin','Ticket',"PassengerId", "Name"], axis = 1)

eda_train.dropna(axis = 0, inplace = True)

eda_train.to_csv('/content/drive/MyDrive/Colab Notebooks/Machine Learning Projects/Titanic/train_eda.csv', index=False, header=True) #export for eda

eda_train.info()
```

```python
#train ready

dummies_train = pd.get_dummies(eda_train, columns=['Pclass', "Sex", "Embarked"])

dummies_train.to_csv("/content/drive/MyDrive/Colab Notebooks/Machine Learning Projects/Titanic/train_data.csv", index=False, header=True) #export for testing
```

```python
## fill null for test
null_class = test.at[test.loc[pd.isnull(test['Fare'])].index[0], "Pclass"]

#print(null_class)
test.loc[pd.isnull(test['Fare']) & (test['Pclass'] == test.at[test.loc[pd.isnull(test['Fare'])].index[0], "Pclass"]), "Fare"] = test.groupby(['Pclass'])['Fare'].mean()[null_class]

#drop columns in test
test.drop(['PassengerId',"Name",'Ticket','Cabin'], axis = 1, inplace = True)

dummies_test = pd.get_dummies(test, columns=['Pclass', "Sex", "Embarked"])

dummies_test.to_csv("/content/drive/MyDrive/Colab Notebooks/Machine Learning Projects/Titanic/test_data.csv", index=False, header=True) #export for testing
```
